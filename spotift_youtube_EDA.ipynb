```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\sahil\Desktop\New folder\Spotify_Youtube.csv')

print(df.head())
```

```python
df
```

```python
df = df.drop(['Url_spotify', 'Url_youtube'], axis=1)
```

```python
df = df.drop(['Uri'], axis=1)
```

```python
df
```

```python
# 'Views' sütununda en yüksek 10 değeri içeren veriyi seçin
top_10_views = df.nlargest(10, 'Views')

# Şarkı adı, sanatçı adı, izlenme sayısı, beğeni sayısı ve yorum sayısı bilgilerini alın
top_10_data = top_10_views[['Track', 'Artist', 'Views', 'Likes', 'Comments']]

# Sonucu görüntüleyin
print(top_10_data)
```

```python
import matplotlib.pyplot as plt

top_10_views = df.nlargest(10, 'Views')

song_names = top_10_views['Track']
view_counts = top_10_views['Views']

plt.figure(figsize=(10, 6))
plt.bar(song_names, view_counts, color='skyblue')
plt.xlabel('Şarkı Adı')
plt.ylabel('İzlenme Sayısı')
plt.title('En Çok İzlenen 10 Şarkı ve İzlenme Sayıları')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_views = df.nlargest(10, 'Views')

song_names = top_10_views['Track']
view_counts = top_10_views['Views']

plt.figure(figsize=(10, 6))
bars = plt.bar(song_names, view_counts, color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

plt.xlabel('Şarkı Adı')
plt.ylabel('İzlenme Sayısı')
plt.title('En Çok İzlenen 10 Şarkı ve İzlenme Sayıları')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_likes = df.nlargest(10, 'Likes')

song_names = top_10_likes['Track']
like_counts = top_10_likes['Likes']

plt.figure(figsize=(10, 6))
bars = plt.bar(song_names, like_counts, color='lightgreen')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

plt.xlabel('Şarkı Adı')
plt.ylabel('Beğeni Sayısı')
plt.title('En Çok Beğeni Alan 10 Şarkı ve Beğeni Sayıları')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_comments = df.nlargest(10, 'Comments')

song_names = top_10_comments['Track']
comment_counts = top_10_comments['Comments']

plt.figure(figsize=(10, 6))
bars = plt.bar(song_names, comment_counts, color='salmon')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

plt.xlabel('Şarkı Adı')
plt.ylabel('Yorum Sayısı')
plt.title('En Çok Yorum Yapılan 10 Şarkı ve Yorum Sayıları')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_views = df.nlargest(100, 'Views')

danceability_values = top_10_views['Danceability']

view_counts = top_10_views['Views']

plt.figure(figsize=(10, 6))
plt.scatter(danceability_values, view_counts, color='orange', alpha=0.7)
plt.title('Danceability Değeri ve Dinlenme Sayısı Arasındaki İlişki')
plt.xlabel('Danceability')
plt.ylabel('Dinlenme Sayısı')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_10_views = df.nlargest(100, 'Views')

like_counts = top_10_views['Likes']
energy_values = top_10_views['Energy']

plt.figure(figsize=(10, 6))
sns.scatterplot(x=like_counts, y=energy_values, color='blue')
plt.title('Beğeni Sayıları ile Energy Değeri Arasındaki İlişki')
plt.xlabel('Beğeni Sayıları')
plt.ylabel('Energy Değeri')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(like_counts, color='green', kde=True)
plt.title('Beğeni Sayıları Dağılımı')
plt.xlabel('Beğeni Sayıları')
plt.ylabel('Frekans')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(energy_values, color='orange', kde=True)
plt.title('Energy Değeri Dağılımı')
plt.xlabel('Energy Değeri')
plt.ylabel('Frekans')
plt.grid(True)
plt.tight_layout()
plt.show()
```
